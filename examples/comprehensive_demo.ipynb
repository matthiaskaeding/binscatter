# Binscatter: Comprehensive Feature Demo

This notebook demonstrates all features of the `binscatter` library using readily available datasets.

## What is Binscatter?

Binscatter plots group the x-axis into bins and plot average outcomes for each bin, providing a clearer view of the relationship between two variables. This is particularly useful when:
- You have large datasets with overplotting
- You want to visualize conditional means
- You need to control for confounders

This implementation follows Cattaneo et al. (2024), "On Binscatter," *American Economic Review*.

```python
import plotly.express as px
import pandas as pd
import numpy as np
from binscatter import binscatter

# Load sample datasets
gapminder = px.data.gapminder()
iris = px.data.iris()
tips = px.data.tips()

print(f"Gapminder: {gapminder.shape[0]} rows")
print(f"Iris: {iris.shape[0]} rows")
print(f"Tips: {tips.shape[0]} rows")
```

## 1. Basic Binscatter Plot

Let's start with a simple example: GDP per capita vs. Life Expectancy from the Gapminder dataset.

```python
# Basic binscatter plot
fig = binscatter(gapminder, x="gdpPercap", y="lifeExp")
fig.update_layout(title="GDP per Capita vs Life Expectancy (Default DPI selector)")
fig.show()
```

## 2. Automatic Bin Selection Methods

The library supports two automatic bin selection methods:
- **DPI (Direct Plug-In)**: Default method, provides optimal bin count
- **ROT (Rule-of-Thumb)**: Alternative automatic selector

```python
# DPI selector (default)
fig_dpi = binscatter(gapminder, x="gdpPercap", y="lifeExp", num_bins="dpi")
fig_dpi.update_layout(title="DPI Bin Selector")
fig_dpi.show()
```

```python
# Rule-of-thumb selector
fig_rot = binscatter(gapminder, x="gdpPercap", y="lifeExp", num_bins="rot")
fig_rot.update_layout(title="Rule-of-Thumb Bin Selector")
fig_rot.show()
```

## 3. Manual Bin Specification

You can manually specify the number of bins for more granular or coarser views.

```python
# Coarse binning (10 bins)
fig_10 = binscatter(gapminder, x="gdpPercap", y="lifeExp", num_bins=10)
fig_10.update_layout(title="10 Bins (Coarse)")
fig_10.show()
```

```python
# Fine binning (50 bins)
fig_50 = binscatter(gapminder, x="gdpPercap", y="lifeExp", num_bins=50)
fig_50.update_layout(title="50 Bins (Fine)")
fig_50.show()
```

```python
# Very fine binning (120 bins)
fig_120 = binscatter(gapminder, x="gdpPercap", y="lifeExp", num_bins=120)
fig_120.update_layout(title="120 Bins (Very Fine)")
fig_120.show()
```

## 4. Controlling for Confounders

One of the most powerful features is the ability to partial out control variables. This shows the relationship between x and y after accounting for other variables.

```python
# Without controls
fig_no_ctrl = binscatter(gapminder, x="gdpPercap", y="lifeExp")
fig_no_ctrl.update_layout(title="Without Controls")
fig_no_ctrl.show()
```

```python
# With single control (year)
fig_ctrl_year = binscatter(gapminder, x="gdpPercap", y="lifeExp", controls="year")
fig_ctrl_year.update_layout(title="Controlling for Year")
fig_ctrl_year.show()
```

```python
# With multiple controls (year and population)
# Note: We use num_bins=20 to avoid issues with automatic selection when pop (large values) is a control
fig_ctrl_multi = binscatter(gapminder, x="gdpPercap", y="lifeExp", controls=["year", "pop"], num_bins=20)
fig_ctrl_multi.update_layout(title="Controlling for Year and Population")
fig_ctrl_multi.show()
```

### Categorical Controls

The library also handles categorical controls automatically via one-hot encoding.

```python
# Control for continent (categorical)
fig_ctrl_cat = binscatter(gapminder, x="gdpPercap", y="lifeExp", controls="continent")
fig_ctrl_cat.update_layout(title="Controlling for Continent (Categorical)")
fig_ctrl_cat.show()
```

```python
# Mix of numeric and categorical controls
fig_ctrl_mixed = binscatter(gapminder, x="gdpPercap", y="lifeExp", controls=["year", "continent"], num_bins=20)
fig_ctrl_mixed.update_layout(title="Controlling for Year and Continent")
fig_ctrl_mixed.show()
```

## 5. Polynomial Overlay

Add polynomial fit lines to visualize the underlying trend. Supports degrees 1 (linear), 2 (quadratic), and 3 (cubic).

```python
# Linear fit overlay
fig_poly1 = binscatter(gapminder, x="gdpPercap", y="lifeExp", poly_line=1)
fig_poly1.update_layout(title="With Linear Fit Overlay")
fig_poly1.show()
```

```python
# Quadratic fit overlay
fig_poly2 = binscatter(gapminder, x="gdpPercap", y="lifeExp", poly_line=2)
fig_poly2.update_layout(title="With Quadratic Fit Overlay")
fig_poly2.show()
```

```python
# Cubic fit overlay
fig_poly3 = binscatter(gapminder, x="gdpPercap", y="lifeExp", poly_line=3)
fig_poly3.update_layout(title="With Cubic Fit Overlay")
fig_poly3.show()
```

### Polynomial with Controls

You can combine polynomial overlays with controls for even more sophisticated analysis.

```python
# Polynomial fit after controlling for year
fig_poly_ctrl = binscatter(
    gapminder, 
    x="gdpPercap", 
    y="lifeExp", 
    controls="year",
    poly_line=2,
    num_bins=20
)
fig_poly_ctrl.update_layout(title="Quadratic Fit with Year Control")
fig_poly_ctrl.show()
```

## 6. Return Types

The function can return either a Plotly figure (default) or a native dataframe with the binned data.

```python
# Return as Plotly figure (default)
fig = binscatter(gapminder, x="gdpPercap", y="lifeExp", return_type="plotly")
print(f"Type: {type(fig)}")
fig.show()
```

```python
# Return as native dataframe
df_binned = binscatter(gapminder, x="gdpPercap", y="lifeExp", return_type="native")
print(f"Type: {type(df_binned)}")
print(f"\nBinned data (first 10 rows):")
df_binned.head(10)
```

## 7. Custom Plotly Arguments

You can pass additional keyword arguments to customize the Plotly scatter plot appearance.

```python
# Custom colors
fig_custom = binscatter(
    gapminder, 
    x="gdpPercap", 
    y="lifeExp",
    color_discrete_sequence=["red"],
)
fig_custom.update_layout(title="Custom Red Color")
fig_custom.show()
```

```python
# Custom labels and title
fig_labels = binscatter(
    gapminder, 
    x="gdpPercap", 
    y="lifeExp",
    title="Economic Development and Longevity",
)
fig_labels.update_xaxes(title="GDP per Capita (USD)")
fig_labels.update_yaxes(title="Life Expectancy (years)")
fig_labels.show()
```

## 9. Backend Compatibility

The library works with multiple dataframe backends thanks to the narwhals abstraction layer. Let's demonstrate with Polars (if available).

```python
try:
    import polars as pl
    
    # Convert to Polars DataFrame
    gapminder_pl = pl.from_pandas(gapminder)
    
    # Use with Polars
    fig_polars = binscatter(gapminder_pl, x="gdpPercap", y="lifeExp")
    fig_polars.update_layout(title="Binscatter with Polars DataFrame")
    fig_polars.show()
    
    print("✓ Polars backend working!")
except ImportError:
    print("Polars not installed - skipping Polars demo")
    print("Install with: pip install polars")
```

```python
try:
    import duckdb
    
    # Create DuckDB connection and relation
    con = duckdb.connect()
    gapminder_duck = con.from_df(gapminder)
    
    # Use with DuckDB
    fig_duck = binscatter(gapminder_duck, x="gdpPercap", y="lifeExp")
    fig_duck.update_layout(title="Binscatter with DuckDB")
    fig_duck.show()
    
    print("✓ DuckDB backend working!")
except ImportError:
    print("DuckDB not installed - skipping DuckDB demo")
    print("Install with: pip install duckdb")
```

## 10. Advanced Example: Complete Analysis

Let's put it all together with a comprehensive analysis showing the impact of different parameters.

```python
# Create a synthetic dataset with a clear relationship
np.random.seed(42)
n = 5000

df_synthetic = pd.DataFrame({
    'x': np.random.uniform(0, 10, n),
    'confounder1': np.random.normal(0, 2, n),
    'confounder2': np.random.normal(5, 1, n),
    'category': np.random.choice(['A', 'B', 'C'], n)
})

# Generate y with confounders
df_synthetic['y'] = (
    3 * np.sin(df_synthetic['x']) +  # Non-linear relationship with x
    0.5 * df_synthetic['confounder1'] +  # Linear relationship with confounder1
    0.3 * df_synthetic['confounder2'] +  # Linear relationship with confounder2
    np.random.normal(0, 0.5, n)  # Noise
)

print(f"Synthetic dataset: {df_synthetic.shape[0]} rows")
df_synthetic.head()
```

```python
# Without controls (confounded)
fig_syn_raw = binscatter(
    df_synthetic, 
    x="x", 
    y="y",
    num_bins=30,
    poly_line=3
)
fig_syn_raw.update_layout(title="Synthetic Data: Raw Relationship (Confounded)")
fig_syn_raw.show()
```

```python
# With numeric controls (deconfounded)
fig_syn_ctrl = binscatter(
    df_synthetic, 
    x="x", 
    y="y",
    controls=["confounder1", "confounder2"],
    num_bins=30,
    poly_line=3
)
fig_syn_ctrl.update_layout(title="Synthetic Data: After Controlling for Confounders")
fig_syn_ctrl.show()
```

```python
# With all controls (numeric + categorical)
fig_syn_all = binscatter(
    df_synthetic, 
    x="x", 
    y="y",
    controls=["confounder1", "confounder2", "category"],
    num_bins=30,
    poly_line=3
)
fig_syn_all.update_layout(title="Synthetic Data: All Controls (Numeric + Categorical)")
fig_syn_all.show()
```
